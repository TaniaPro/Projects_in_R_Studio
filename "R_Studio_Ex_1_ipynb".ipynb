 **TASK 1:** **Calculate a correlation between budget and total gross, for films with a reported budget, without War films. Print the correlation.**


```python
#read files
setwd("D:\\My files\\Study\\סמסטר ג\\R\\A3")

imdb.movies<-read.csv("IMDB_movies.csv")
head(imdb.movies)

player.df<-read.csv("IMDB_players.csv")
head(player.df)

#Budget to numeric
summary(imdb.movies$budget)
imdb.movies$budget<-as.numeric(imdb.movies$budget)
class(imdb.movies$budget)

#Data Imputation-Total Gross
summary(imdb.movies)
imdb.movies$total.gross[is.na(imdb.movies$total.gross)] <- 
  mean(imdb.movies$total.gross, na.rm = TRUE) 

#remove N/A from budget
imdb.movies.budget<-imdb.movies[is.na(imdb.movies$budget)==FALSE,]


#No war genre
nowar<-imdb.movies.budget[imdb.movies.budget$genre!="War",]


#correlation between budget and total gross
cor(nowar$budget,nowar$total.gross,method=c("pearson"))
```

**TASK 3: Are there differences (on average) between total gross movie profits that have a budget report and those that do not have.**


1.   Run test t and print the output of the test result.
2.   Explain in a few words what the t test is and what is it's purpose .
3. Answer: What is the conclusion of the test?

```python
#TASK 3
#Total gross by Budget
a<-imdb.movies[is.na(imdb.movies$budget)==FALSE,]
b<-imdb.movies[is.na(imdb.movies$budget)==TRUE,]

total.gross.budget<-a$total.gross
total.gross.no.budget<-b$total.gross

#T-Test
t.test(total.gross.budget,total.gross.no.budget, 
       mu=0, alt='two.sided', conf=0.95, var.eq=F, paired=F)
```

**Estimated Output (1):**
Capture.PNG

**Estimated Output (2):**

The T test is a statistical test, designed to test hypotheses about the mean, when the variance in the population is unknown. In the task, a test was performed on two groups that are not dependent on each other:
1. Films that have a budget report (vector of total.gross.budget values)
2. Films that do not have a budget report (vector of values ​​total.gross.no.budget).

Since these are independent groups, the difference (on average) between the film profits of the two groups is examined.
H_0 - Hypothesis 0, there is no difference between two measured groups.
H_1- There is a difference between two measured groups.

The significance of the result is measured by the P value. Before performing a test the significance level (α) is determined, it is usually equal to 5%. If the value P is less than or equal to α, the 0 hypothesis is incorrect.
In the formula of t.test place the values ​​X and Y:
X-vector values ​​total.gross.budget
Y-vector values ​​total.gross.no.budget


**Estimated Output (3):**

In the results: a P value  is smaller than the significance level α, this indicates that the 0 hypothesis is incorrect - there are differences (on average) between film profits for which there is a budget report and those that do not have budget report. The averages of two groups are not equal to each other.
